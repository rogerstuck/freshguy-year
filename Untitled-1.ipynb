```python
#全流程
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset,DataLoader
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

#设置随机种子,保证结果可复现
torch.manual_seed(42)
np.random.seed(42)

#检查是否有gpu
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# 加载数据
columns = ['user_id', 'item_id', 'rating', 'timestamp']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=columns)

# 对用户ID和电影ID进行编码（从0开始）
user_encoder = LabelEncoder()
item_encoder = LabelEncoder()

ratings['user_idx'] = user_encoder.fit_transform(ratings['user_id'])
ratings['item_idx'] = item_encoder.fit_transform(ratings['item_id'])

# 获取用户数和物品数
n_users = ratings['user_idx'].nunique()
n_items = ratings['item_idx'].nunique()

print(f"用户数: {n_users}")
print(f"电影数: {n_items}")
print(f"评分数量: {len(ratings)}")

# 划分训练集和测试集
train_data, test_data = train_test_split(ratings, test_size=0.2, random_state=42)
print(f"训练集: {len(train_data)} 条")
print(f"测试集: {len(test_data)} 条")

#创建数据集类
class RatingDataset(Dataset):
    def __init__(self,data):
        self.users = torch.LongTensor(data['user_idx'].values)
        self.items = torch.LongTensor(data['item_idx'].values)
        self.ratings = torch.FloatTensor(data['rating'].values)
    
    def __len__(self):
        return len(self.ratings)

    def __getitem___(self,idx):
        return {
            'user':self.users[idx],
            'item':self.items[idx],
            'rating':self.ratings[idx]
        }

#创建数据加载器
train_dataset = RatingDataset(train_data)
test_dataset = RatingDataset(test_data)
train_loader = DataLoader(train_dataset,batch_size=256,shuffle=True)
#数据集为train_dataset,批次大小为256,每个epoch(训练周期)打乱顺序
test_loader = DataLoader(test_dataset,batch=256,shuffle=False)

#定义简单的矩阵分解模型
class MatrixFactorization(nn.Module):
    """矩阵分解模型"""
    def __init__(self,n_users,n_items,n_factors=50):
        """
        参数:
            n_users: 用户数量
            n_items: 物品数量
            n_factors: 潜在因子数量（向量维度）
        """
        super(MatrixFactorization,self).__init__()

        #用户和物品的嵌入层
        self.user_factors = nn.Embedding(n_users,n_factors)
        self.item_factors = nn.Embedding(n_items,n_factors)
        #将每个用户/物品ID映射为一个n_factors维的向量,作为模型要学习的"隐向量"

        #用户和物品的偏置项
        self.user_bias = nn.Embedding(n_users,1)
        self.item_bias = nn.Embedding(n_items,1)
        #偏置项用于捕捉用户或物品自身的评分倾向（比如有些用户习惯给高分）

        #初始化权重
        self._init_weights()

    def _init_weights(self):
        """初始化模型权重"""
        #使用正态分布初始化嵌入层
        nn.init.normal_(self.user_factors.weight,std=0.01)
        nn.init.normal_(self.item_foctors.weight,std=0.01)
        #用均值为0、标准差0.01的正态分布初始化用户和物品的隐向量
        #小随机数有助于打破对称性，避免梯度消失/爆炸

        #偏置项初始化为0,让模型从0开始学习偏差
        nn.init.zeros_(self.user_bias.weight)
        nn.init.zeros_(self.item_bias.weight)

    def forward(self,user,item):
        """
        前向传播：预测评分
        参数:
            user: 用户索引
            item: 物品索引
        返回:
            预测评分 (范围: 1-5)
        """
        #计算用户向量和物品向量的点积
        pred = (self.user_factors(user)*self.item_factors(item).sum(1,keepdim=True))
        #这里只有两个维度:batch_size和n_factor,0和1
        """
        e.g.
        # user_factors(user) 输出（2个用户，每个3维）
        user_vec = [[0.1, 0.2, 0.3],  # 用户1的向量
                    [0.4, 0.5, 0.6]]  # 用户2的向量

        # item_factors(item) 输出（2个物品，每个3维）
        item_vec = [[0.7, 0.8, 0.9],  # 物品1的向量
                    [0.2, 0.3, 0.4]]  # 物品2的向量

        # 元素级相乘
        product = user_vec * item_vec = [[0.1*0.7, 0.2*0.8, 0.3*0.9],  = [[0.07, 0.16, 0.27],
                                          [0.4*0.2, 0.5*0.3, 0.6*0.4]]    [0.08, 0.15, 0.24]]

        # 求和（得到点积）
        pred = product.sum(1, keepdim=True) = [[0.07+0.16+0.27], = [[0.50],
                                                [0.08+0.15+0.24]]    [0.47]]
        """
        pred += self.user_bias(user) + self.item_bias(item)
        #从嵌入层获取用户和物品的向量
        #计算对应元素的乘积，然后在维度1上求和（得到点积）
        #keepdim=True保持输出维度为(batch_size, 1)

        #将预测值限制在1-5之间
        #使用sigmoid映射到0-1,然后缩放到1-5
        return torch.sigmoid(pred) * 4 + 1

    def predict(self,user,item):
        #封装的预测方法，用于推理阶段
        """预测单个用户-物品对"""
        self.eval() #eval()切换到评估模式（影响Dropout等层）
        with torch.no_grad(): #torch.no_grad()禁用梯度计算，节省内存和计算
            return self.forward(user,item) #直接调用forward进行预测

#初始化模型
n_factors = 50
model = MatrixFactorization(n_users,n_items,n_factors).to(device)

#查看模型结构
print(model)
print(f"模型参数总数:{sum(p.numel() for p in model.parameters()):,}")

#定义训练函数
def train_epoch(model,train_loader,optimizer,criterion,device):
    """训练一个epoch"""
    model.train() #将模型设置为训练模式(启用Dropout、BatchNorm等训练特性)
    total_loss = 0 #初始化总损失累加器

    for batch in train_loader: #遍历训练数据加载器中的每个批次
        #将数据转移到指定设备
        users = batch['user'].to(device)
        items = batch['item'].to(device)
        ratings = batch['rating'].to(device)

        #前向传播
        optimizer.zero_grad() #清零梯度缓冲区,避免梯度累积
        predictions = model(users,items).squeeze() #模型预测,squeeze()去除多余的维
        loss = criterion(predictions,ratings)

        #反向传播
        loss.backward()
        optimizer.step() # 更新模型参数

        total_loss += loss.item()

    return total_loss/len(train_loader) #返回平均损失

def evaluate(model,test_loader,criterion,device):
    """评估模型"""
    model.eval() # 将模型设置为评估模式(禁用Dropout,使用固定的BatchNorm参数)
    total_loss = 0
    all_predictions = [] #存储所有预测值的列表
    all_ratings = [] # 存储所有真实评分的列表

    with torch.no_grad(): # 禁用梯度计算，减少内存消耗并加速计算
        for batch in test_loader:
            users = batch['user'].to(device)
            items = batch['item'].to(device)
            ratings_batch = batch['rating'].to(device)
            predictions = model(users, items).squeeze()
            loss = criterion(predictions, ratings_batch)
            total_loss += loss.item()
            #将预测值和真实值移到CPU并转换为numpy数组存储
            all_predictions.extend(predictions.cpu().numpy())
            all_ratings.extend(ratings_batch.cpu().numpy())
    
    #计算RMSE(均方根误差)
    mse = torch.loss/len(test_loader)
    rmse = np.sqrt(mse)

    return mse,rmse,np.array(all_predictions),np.array(all_ratings) # 返回评估指标与预测结果

#训练模型
#定义损失函数与优化器
criterion = nn.MSELoss() #均方误差损失
optimizer = optim.Adam(model.parameters(),lr=0.01,weight_decay=1e-5)
"""
创建Adam优化器，负责更新模型参数
model.parameters(): 获取模型所有可训练参数
lr=0.01: 学习率，控制参数更新的步长
weight_decay=1e-5: L2正则化系数，防止过拟合
"""

#训练参数
n_epochs = 20 #训练轮数
train_losses = []
test_losses = []
test_rmses = []

for epoch in range(n_epochs):
    train_loss = train_epoch(model,train_loader,optimizer,criterion,device)
    #调用之前定义的train_epoch训练函数进行一轮训练
    #返回该epoch的平均训练损失
    train_losses.append(train_loss) #将训练损失添加到列表中,用于后续绘图

    #评估
    test_loss,test_rmse,_,_ = evaluate(model,test_loader,criterion,device)
    #调用evaluate评估函数在测试集上评估模型性能
    #返回测试损失、RMSE，以及预测值和真实值（这里用_忽略后两个返回值）

    test_losses.append(test_loss)
    test_rmses.append(test_rmse)

    # 打印当前epoch的训练结果
    print(f"Epoch {epoch+1:2d}/{n_epochs} | "  # epoch+1: 从1开始计数，:2d表示占2位宽度
          f"训练损失: {train_loss:.4f} | "      # 训练损失保留4位小数
          f"测试损失: {test_loss:.4f} | "       # 测试损失保留4位小数
          f"RMSE: {test_rmse:.4f}")             # RMSE保留4位小数

#训练结束部分
print("="*60)  # 打印分隔线
print(f"训练完成！最终RMSE: {test_rmses[-1]:.4f}")
# 输出训练完成信息，并显示最后一个epoch的RMSE
# test_rmses[-1] 获取列表中的最后一个元素（最终RMSE）

#可视化训练过程
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4)) #创建新图形窗口,大小为12英寸宽,4英寸高

#损失曲线
plt.subplot(1,2,1)
#创建1行2列的子图布局,并选中一个子图
    #(1,2,1)表示:总共1行2列,当前激活第一个子图
#绘制训练损失曲线
plt.plot(range(1,n_epochs+1),train_losses,'b-',label='训练损失',linewidth=2)
    # range(1, n_epochs+1): X轴数据，从1到epoch总数（排除0）
    # train_losses: Y轴数据，每个epoch的训练损失值
    # 'b-': 蓝色(b)实线(-)样式
    # label='训练损失': 图例标签
    # linewidth=2: 线宽为2像素
#绘制测试损失曲线,‘r-’红色实线
plt.plot(range(1, n_epochs+1), test_losses, 'r-', label='测试损失', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.title('训练过程损失曲线')
plt.legend() # 显示图例（根据之前设置的label）
plt.grid(True, alpha=0.3)# 显示网格线
# True: 启用网格
# alpha=0.3: 网格线透明度为0.3（0为完全透明，1为完全不透明）

#RMSE曲线
plt.subplot(1, 2, 2)
plt.plot(range(1, n_epochs+1), test_rmses, 'g-', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('RMSE')
plt.title('测试集RMSE')
plt.grid(True, alpha=0.3)

plt.tight_layout()# 自动调整子图参数，使子图之间以及子图与边框之间有合适的间距
# 防止标签重叠，优化布局
plt.show()# 显示图形窗口，呈现所有绘图内容
# 如果没有这行代码，图形可能不会显示
```
